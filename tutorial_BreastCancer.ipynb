Pandas é utilizado para importar dados.

```
import pandas as pd

from google.colab import files
myFile = files.upload()
x_train = pd.read_csv("xtrain.csv", header=None)
y_train = pd.read_csv("ytrain.csv", header=None)
x_test = pd.read_csv("xtest.csv", header=None)
y_test = pd.read_csv("ytest.csv", header=None)
```

Definir uma rede neural (classifier) com modelo sequencial de Keras com base em dados de treino. Esta rede neural, pode-se prever resultados futuros do modelo.

```
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential() #inicia a rede neural

#primeiro nível da rede - 16 neurônios com 30 entradas e ativação da saída 'relu'
classifier.add(Dense(units = 16, activation = 'relu', input_dim = 30))
#segundo nível da rede - 8 neurônios com 16 entradas e ativação da saída 'relu'
classifier.add(Dense(units = 8, activation = 'relu'))
#terceiro nível da rede - 6 neurônios com 8 entradas e ativação da saída 'relu'
classifier.add(Dense(units = 6, activation = 'relu'))
#quarto nível da rede - 1 neurônios com 6 entradas e ativação da saída 'sigmoid'
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

Quando a rede estiver concluida, deve-se especificar o otimizador e a função loss custo?)

```
classifier.compile(optimizer='rmsprop', loss = 'binary_crossentropy')
```

O próximo passo é treinar a rede neural classifier com os dados de treino x e y, de tal forma que a rede irá abstrair uma relação entre eles e construir uma rede de neurônios que possa replicá-los.



```
classifier.fit(x_train,y_train,batch_size=1,epochs=100)
```

Loss é uma medida de acurácia da rede neural?

Para prever novos valores de y em termos de probabilidade, indicamos um conjunto de teste x_test que a rede não viu ainda. Com os valores de y da previsão, fazemso uma limpeza, dizendo que os valores acima 0.5 se considerados como 1 e os valores abaixo, 0.

```
y_pred = classifier.predict(x_test) # previsão em cima dos dados de teste.
y_pred_n = (1 if y>= 0.5 else 0 for y in y_pred) #limpeza - generator object
y_pred_n = list(y_pred_n) # list - meu primeiro problema resolvido devido a inconsistência de versão

#plotar Y_pred e y_pred_n
```

Comparamos y_pred_n com y_test para avaliar o desempenho da rede classifier.

```
total = 0
correct = 0
wrong = 0
for i in y_pred_n:
  total = total+1
  if(y_test.at[i,0] == y_pred_n[i]):
    correct=correct+1
  else:
    wrong=wrong+1

print("Total " + str(total))
print("Acertos " + str(correct))
print("Erros " + str(wrong))
```
