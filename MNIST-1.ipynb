CNN MNIST com 11,2% de acurácia reduzindo para 10 o número de neurônio na primeira camada Dense.

```
import tensorflow as tf
from keras.datasets import mnist
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.models import Sequential
```

```
batch_size = 128
num_classes = 10
epochs = 10
```

```
(x_train, y_train), (x_test, y_test) = mnist.load_data() #formato do dataset
```

```
print(x_train.shape)
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(x_test.shape[0],x_test.shape[1],x_test.shape[2],1)
print(x_train.shape)
```

```
model = Sequential() # modelo sequencial
model.add(Conv2D(64, kernel_size=(5,5), strides=(1,1), # primeira camada Convolicional
                 activation='relu',
                 input_shape = (28,28,1))) # largura, altura, canais
model.add(MaxPooling2D(pool_size=(5,5),strides=(2,2))) # reduz o tamanho da imagem com pixels mais relevantes
model.add(Conv2D(32,(5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Flatten()) # transforma em vetor
model.add(Dense(10,activation='relu')) # todas as entradas são conectadas ao 1000 neurônios
model.add(Dense(num_classes, activation='softmax')) # 100 entradas conectadas em num_classes neurônios

model.summary()
```

```
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(x=x_train,y=y_train,epochs=epochs) # ajuste dos dados de treino à rede neural
score = model.evaluate(x_test,y_test,verbose=0) # performance da rede
```
